```python
import pandas as pd
N_A=pd.read_csv("Nd.csv") 
N_A
```

```python
                            #Distribution of Rating Category(Univariate Analysis)


import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.countplot(data=N_A, x='rating_category')
plt.title('Distribution of Rating Category')
plt.xticks(rotation=45)
plt.show()
```

```python
                               #Distribution of Duration_Category(Univariate Analysis)

plt.figure(figsize=(10,6))
sns.countplot(data=N_A, x='Duration_Category')
plt.title('Distribution of Duration Category')
plt.xticks(rotation=45)
plt.show()
```

```python
                                #Distribution of Types(Movies or Series)(Univariate Analysis)

plt.figure(figsize=(10,6))
sns.countplot(data=N_A, x='type')
plt.title('Distribution of types')
plt.xticks(rotation=45)
plt.show()
```

```python
                                       #STACKED BAR PLOT (Bivariate Analysis)


country_rating = pd.crosstab(N_A['country_category'], N_A['rating_category'])
country_rating.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Rating Category by Country Category')
plt.xlabel('Country Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
                                         #STACKED BAR PLOT (Bivariate Analysis)

country_rating = pd.crosstab(N_A['movie_age'], N_A['rating_category'])
country_rating.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Rating Category by  movie_age')
plt.xlabel('Movie age')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
                                         #STACKED BAR PLOT (Bivariate Analysis)

country_rating = pd.crosstab(N_A['Content Age'], N_A['rating_category'])
country_rating.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Rating Category by Content Age')
plt.xlabel('Content_Age')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
                                        #HEATMAP(Correlation Analysis)

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

rating_country_crosstab = pd.crosstab(N_A['rating_category'], N_A['country_category'])

plt.figure(figsize=(10, 6))
sns.heatmap(rating_country_crosstab, annot=True, cmap='Blues', fmt='d', cbar=True)

plt.title("Heatmap of Rating Category vs Country Category")
plt.xlabel("Country Category")
plt.ylabel("Rating Category")
plt.show()
```

```python
                                       #HEATMAP(Correlation Analysis)


import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

rating_country_crosstab = pd.crosstab(N_A['rating_category'], N_A['Title_Words'])

plt.figure(figsize=(10, 6))
sns.heatmap(rating_country_crosstab, annot=True, cmap='Blues', fmt='d', cbar=True)

plt.title("Heatmap of Rating Category vs Movie Title Words")
plt.xlabel("Title words Category")
plt.ylabel("Rating Category")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder 

Categorical_columns=["type","rating_category","Duration_Category","country_category","Directors","Title_Words","Content Age"] 

le=LabelEncoder() 

for col in Categorical_columns:
    N_A[col]=le.fit_transform(N_A[col]) 

N_A.head()
```

```python
from sklearn.preprocessing import StandardScaler 

scaler=StandardScaler() 

N_A["movie_age"]=scaler.fit_transform(N_A["movie_age"].values.reshape(-1, 1)) 

N_A.head()
```

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.model_selection import train_test_split

X=N_A.drop(columns=["rating_category"]) 
y=N_A["rating_category"]  

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42) 

imp_rf=RandomForestClassifier(n_estimators=200,random_state=42) 

imp_rf.fit(X_train,y_train)

importances=imp_rf.feature_importances_

feature_importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': importances
})

feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)


feature_importance_df

```
